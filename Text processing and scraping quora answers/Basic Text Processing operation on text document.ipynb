```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import *
from nltk.stem.wordnet import WordNetLemmatizer
```

# Reading json file

```python
import json
```

```python
with open("quora.json", "r") as file:
    data=json.load(file)
ans=[]
for i in data["answer"]:
    ans.append([i['answer']])
```

# Text processing

```python
def converttolower(text):
    return text.lower()
for i in range(5):
    ans[i]=[converttolower(ans[i][0])]
ans
```

```python
import string
import re
```

```python
def removeurl(text):
    return re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
for i in range(5):
    ans[i]=[removeurl(ans[i][0])]
ans
```

```python
def removepunctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))
for i in range(5):
    ans[i]=[removepunctuation(ans[i][0])]
ans
```

```python
import nltk
```

```python
def tokenization(text):
    return nltk.word_tokenize(text)
for i in range(5):
    ans[i]=[tokenization(ans[i][0])]
ans
```

```python
def removestop(text):
    return [word for word in text if not word in stopwords.words()]
for i in range(5):
    ans[i]=[removestop(ans[i][0])]
ans
```

```python
porter = PorterStemmer()
lematizer=WordNetLemmatizer()
def stemming(text):
    [porter.stem(word) for word in text]
for i in range(0):
    ans[i]=[stemming(ans[i][0])]
def lematizing(text):
    [lematizer.lemmatize(word) for word in text] 
for i in range(0):
    ans[i]=[lematizing(ans[i][0])]
```

# Creating TDM

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer()
for i in range(5):
    X = vec.fit_transform(ans[i][0])
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df)
```
